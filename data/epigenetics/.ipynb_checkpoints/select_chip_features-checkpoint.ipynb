```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_cell_line_name(feature_names_selected):
    return([feature_name.split('Cell-Line=')[1].split('#')[0] for feature_name in feature_names_selected])

def get_gene_names(names_file_path):
    names_file = open(names_file_path,'r', encoding='ISO-8859-1')

    feature_names = []
    feature_names_full = []
    for line in names_file: 
        line_spl = line.split("/")
#         print(line, line_spl)
#         print(line_spl[3])
        line_names = line_spl[3].split(":")
#         print(line_names[0])

        feature_names.append(line_names[0])
        feature_names_full.append(line)

    feature_names = np.array(feature_names)
    feature_names_full = np.array(feature_names_full)
    feature_names_uniq = np.unique(feature_names)
    
    return feature_names, feature_names_full, feature_names_uniq

# get_features_indxs_first_selected()

def is_smoothedM(full_name):
    return full_name.split(".")[-3] == 'smoothedM'

# feature_names_full[163], is_smoothedM(full_name=feature_names_full[163])

def get_features_indxs_smoothedM(feature_names_uniq, feature_names, feature_names_full):
    ''' 
    return:
        first_uniq_indxs : for each unique feature name it's first index in feature_names, 
        names_first_uniq_indx_dict : dict of (feature_names_uniq, first_uniq_indxs) 
    '''
    first_uniq_indxs = []
    for name_un in feature_names_uniq:
        i = 0
        while i < len(feature_names) and (feature_names[i] != name_un or \
            (feature_names[i] == name_un and not is_smoothedM(feature_names_full[i]))):
            i += 1
        if (i >= len(feature_names)):
            i = -1
        first_uniq_indxs.append(int(i))
        
    names_first_uniq_indx_dict = dict(zip(feature_names_uniq, first_uniq_indxs))

    return first_uniq_indxs, names_first_uniq_indx_dict

def get_features_indxs_smoothedM_last(feature_names_uniq, feature_names, feature_names_full):
    ''' 
    return:
        first_uniq_indxs : for each unique feature name it's first index in feature_names, 
        names_first_uniq_indx_dict : dict of (feature_names_uniq, first_uniq_indxs) 
    '''
    last_uniq_indxs = []
    for name_un in feature_names_uniq:
        feature_idx = -1
        for i in range(len(feature_names)):
            if feature_names[i] == name_un and is_smoothedM(feature_names_full[i]):
                feature_idx = i
        
        last_uniq_indxs.append(int(feature_idx))
        
    names_last_uniq_indx_dict = dict(zip(feature_names_uniq, last_uniq_indxs))

    return last_uniq_indxs, names_last_uniq_indx_dict

def get_features_indxs_smoothedM_all(feature_names_uniq, feature_names, feature_names_full):
    ''' 
    return:
        first_uniq_indxs : for each unique feature name it's first index in feature_names, 
        names_first_uniq_indx_dict : dict of (feature_names_uniq, first_uniq_indxs) 
    '''
    all_uniq_indxs = []
    for name_un in feature_names_uniq:
        print("len(feature_names)", len(feature_names))
        feature_idx = []
        for i in range(len(feature_names)):
            if feature_names[i] == name_un and is_smoothedM(feature_names_full[i]):
                feature_idx.append(i)
#                 print(name_un, i, feature_names_full[i])
#         print(feature_idx)
        all_uniq_indxs.append(feature_idx)
#     print(all_uniq_indxs)

    names_all_uniq_indx_dict = dict(zip(feature_names_uniq, all_uniq_indxs))

    return all_uniq_indxs, names_all_uniq_indx_dict


def get_features_indxs_selected(select_feat_names, 
                                      feature_names_uniq, 
                                      feature_names_full,
                                      feature_names, 
                                      ordered_idx='last',
                                      verbose=True):
    '''
    get select_feat_names columns from data
    '''
    
    select_feat_names = np.array(select_feat_names, dtype='<U23')
    if verbose:
        print('selection names: ', select_feat_names)
    
    # Check if the fiven names of features exist in the list of feature columns that we have
    is_in_feature_cols = []
    for name_i in range(len(select_feat_names)):
        is_in_feature_cols.append(select_feat_names[name_i] in feature_names_uniq)
    is_in_feature_cols
    
    # Save features that exist
    select_feat_names = select_feat_names[is_in_feature_cols]
    
    # Get first idnx of the selected features.
    if ordered_idx == 'first':
        #     first_uniq_indxs_select, names_first_uniq_indx_dict_select = \
        #        get_features_indxs_first(select_feat_names, feature_names)
        first_uniq_indxs_select, names_first_uniq_indx_dict_select = get_features_indxs_smoothedM(
                                                                        select_feat_names,
                                                                        feature_names, 
                                                                        feature_names_full)
    elif ordered_idx == 'last':
        first_uniq_indxs_select, names_first_uniq_indx_dict_select = get_features_indxs_smoothedM_last(
                                                                        select_feat_names,
                                                                        feature_names,
                                                                        feature_names_full)
    elif ordered_idx == 'all':
        first_uniq_indxs_select, names_first_uniq_indx_dict_select = get_features_indxs_smoothedM_all(
                                                                        select_feat_names,
                                                                        feature_names,
                                                                        feature_names_full)

    return first_uniq_indxs_select, names_first_uniq_indx_dict_select, select_feat_names


# get_features_selected_data()

def get_features_selected_data(names_file_path, select_feat_names, data, CL_name, 
                                      ordered_idx='last', verbose=True, add_chr_id=False): 
    '''
    get select_feat_names columns from data
    '''
    feature_names, feature_names_full, feature_names_uniq = get_gene_names(names_file_path)

    
    uniq_indxs_select, names_first_uniq_indx_dict_select, select_feat_names = \
        get_features_indxs_selected(select_feat_names, feature_names_uniq, feature_names_full, feature_names,
                                    ordered_idx, verbose=True)
        
    print(uniq_indxs_select, names_first_uniq_indx_dict_select)

    if ordered_idx=='all':
        
#         all_smoot_indxs_select = np.hstack(names_first_uniq_indx_dict_select.values())
        first_uniq_indxs_select = np.hstack(names_first_uniq_indx_dict_select.values()).astype(int)
    
#         first_uniq_indxs_select = []
#         for f in range(len(uniq_indxs_select)):
#             if len(uniq_indxs_select[f]):
#                 first_uniq_indxs_select.append(uniq_indxs_select[f][0])  # use first
#             else:  
#                 first_uniq_indxs_select.append(-1)
                
    else:
        first_uniq_indxs_select = np.array(uniq_indxs_select)
        
    first_uniq_indxs_select = np.array(first_uniq_indxs_select) 
    first_uniq_indxs_select = first_uniq_indxs_select[first_uniq_indxs_select != -1]
        
    # CHECK Cell Line:
    feature_names_selected = feature_names_full[first_uniq_indxs_select]
    is_correct_CL = np.array([ CL == CL_name for CL in get_cell_line_name(feature_names_selected)])
#     print(len(first_uniq_indxs_select), first_uniq_indxs_select[is_correct_CL], is_correct_CL)  
    
    first_uniq_indxs_select = first_uniq_indxs_select[is_correct_CL] 
#     print(len(first_uniq_indxs_select))
    
    select_feat_names = feature_names[first_uniq_indxs_select]
    feature_names_selected = feature_names_full[first_uniq_indxs_select]
    

    # CORRECT THE INDEXES to the data indexes! +2 from the names
    first_uniq_indxs_select = np.array(first_uniq_indxs_select) + 2
    
    if (add_chr_id):
#         first_uniq_indxs_select.append(0)
#         first_uniq_indxs_select.append(1)

#         print(first_uniq_indxs_select)
        first_uniq_indxs_select = list(np.hstack([[0, 1], first_uniq_indxs_select]))
#         select_feat_names  = list(select_feat_names)
#         select_feat_names.append('chrN')
#         select_feat_names.append('start_coord')
#         select_feat_names  = np.array(select_feat_names)

        select_feat_names = list(np.hstack([['chrN', 'start_coord'], select_feat_names]))

    
    data_uniq_features = data[ data.columns[ first_uniq_indxs_select] ]
    data_uniq_features.head()

    data_uniq_features.columns = select_feat_names
    data_uniq_features.head()

    return data_uniq_features, select_feat_names, names_first_uniq_indx_dict_select, \
            feature_names_full, feature_names_selected, first_uniq_indxs_select
```
